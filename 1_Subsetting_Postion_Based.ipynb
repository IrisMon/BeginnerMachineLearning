**DIFERENCE BETWEEN LOC AND ILOC**

```
sample_df=pd.DataFrame({
    'gender':['M', 'F', 'M', 'M', 'F'],
    'grade' :['A', 'A', 'B', 'B', 'A'],
    'marks' :[22, 21, 22,  12,20],
    'id'    :['A101','A102','A103','A104','A105']
})
sample_df
```

Now , when we define a dataframe by default the index is a range of numbers. Let's see what happens if try to slice the dataframe using both loc and iloc

```
sample_df.loc[2:4]
```

```
sample_df.iloc[2:4]
```

* When we try to slice the datafrae using the loc function on range (2 to 4) it first finds out the index with **label 2** and goes till it finds the index with **label 4**

* On the other hand, when we try to slice the dataframe using the **iloc** function on **range(2 to 4)** it starts with the **starting point index 2** and goes till point -1 which is 3

**Let's verify this by sorting the dataframe by marks column**


```
sample_df=sample_df.sort_values(by =['marks'])
sample_df
```

**Now that we have sorted the dataframe, the indices are shuffled accordingly . Let's try to slice the dataframes again.**

```
sample_df.loc[4:2]
```

```
sample_df.iloc[1:4]
```

**TABLE OF CONTENTS**

* How to select rows based on multiple conditions?
* How to select specific columns from a data?
* How to select rows based on a condition and view only the specific columns?
* How to select the columns with specific data types?


```
##View the top rows of the data
data.head(3)
```

**HOW TO SELECT ROWS BASED ON CONDITION?**

```
## Filter rows with condition
# data.loc[data.Outlet_Establishment_Year==1987]
data[data.Outlet_Establishment_Year==1987]
```

```
data.loc[data.Outlet_Establishment_Year==1987]
```

```
data[data['Outlet_Establishment_Year']==1987]
```

**HOW TO SELECT ROWS BASED ON MULTIPLE CONDITIONS?**

* When passing multiple conditions make sure that you put each of the condition in a parenthesis().

* Select all rows where the value of Outlet_Establishment_Year is 1987 and value of Outlet_Size is High

```
#Filter data based on multiple conditions
## Form1: data.loc[(data.Outlet_Establishment_Year==2009) & (data.Outlet_Size == 'Medium')]
# Form2:
data[(data.Outlet_Establishment_Year!=2009)  | (data.Outlet_Size=='Medium')] ## &=And |=O AND NO= ! operator
```

**If you want to filter for a list of values from a column then instead of writing multiple conditions use the** **isin** **function**

**HOW TO FILTER FOR A LIST OF VALUES?**

```
#Get rows for  3 years 1987 , 1988, 1999
data[(data.Outlet_Establishment_Year==1987)  | (data.Outlet_Establishment_Year==1988)  | (data.Outlet_Establishment_Year==1999)] 
```

```
data[data.Outlet_Establishment_Year.isin([1987,1988,1999])]
```

```
data.loc[data.Outlet_Establishment_Year.isin([1987,1988,1999])]
```

**HOW TO SELECT SPECIFIC COLUMNS FROM A DATA?**

* You just need to pass a list of columnhs that you need.

```
##List of columns
select_columns=['Item_Identifier','Item_MRP','Outlet_Establishment_Year','Outlet_Size']

#dataframe with specific columns
data[select_columns]
```

```
select_columns=['Item_Identifier','Item_MRP','Outlet_Establishment_Year','Outlet_Size']

#dataframe with specific columns
data[(data.Outlet_Establishment_Year==1987)&(data.Outlet_Size=='High')][select_columns]
```

**Using loc**

* Using loc , we can provide columns to select within the same square bracket

```
data.loc[(data.Outlet_Establishment_Year== 1987) & (data.Outlet_Size=='High'),select_columns]
```
