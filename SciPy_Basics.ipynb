COMPUTE the Nth Derivae of a function

```
from scipy.misc import derivative
```

```
##Define a function 
def my_function(x):
  return x**2+x+1

# Calculate the fisrt derivate of the function at x=2

derivative (func=my_function , x0=2)
```

PERMUTACION AND COMBINATIONS

```
#COMBINATIONS
from scipy.special import comb
## Total number of combinations from 4 differect values taken 2 at a time
#value of 4C2
com=comb(4,2)
print(com)
```

```
##PERMUTATIONS: Value of 4P2
from scipy.special import perm 
per=perm(4,2)
print(per)
```

LINEAR ALGEBRA

```
from scipy import linalg
import numpy as np
```

```
#square matrix
matrix=np.array([[1,5,2],
                [3,2,1],
                [1,2,1]])
```

```
matrix
```

```
## Pass values to det() function 
linalg.det(matrix) 
```

```
#Inverse of a matrix
linalg.inv(matrix)
```
